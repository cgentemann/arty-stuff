```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import csv

# Function to read the CSV files with the structure shown in your screenshot
def read_light_curve_csv(file_path):
    data = []
    with open(file_path, 'r') as csvfile:
        # Assuming no header row, based on your Excel screenshot
        reader = csv.reader(csvfile)
        for row in reader:
            # Skip empty rows
            if not row or len(row) < 3:
                continue
            try:
                # Extract index, time, and flux from the three columns
                index = int(row[0])
                time = float(row[1])
                flux = float(row[2])
                data.append((index, time, flux))
            except (ValueError, IndexError):
                # Skip rows that don't have the expected format
                continue
    
    # Convert to numpy arrays
    if data:
        indices, times, fluxes = zip(*data)
        return np.array(indices), np.array(times), np.array(fluxes)
    else:
        return np.array([]), np.array([]), np.array([])

# Usage for both files
def plot_star_light_curves(celeano_file, asterope_file):
    # Read data from both files
    cel_idx, cel_time, cel_flux = read_light_curve_csv(celeano_file)
    ast_idx, ast_time, ast_flux = read_light_curve_csv(asterope_file)
    
    # Create figure with two subplots
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 10))
    
    # Plot Celeano
    if len(cel_time) > 0:
        # Normalize time to start from 0
        cel_time_norm = cel_time - cel_time.min()
        
        ax1.plot(cel_time_norm, cel_flux, 'b-', linewidth=1.5, alpha=0.8, label='Celeano')
        ax1.scatter(cel_time_norm, cel_flux, c='red', s=5, alpha=0.7)
        ax1.set_title('Celeano Light Curve', fontsize=14, fontweight='bold')
        ax1.grid(True, alpha=0.3)
        ax1.legend()
        ax1.set_ylabel('Flux (ADU)', fontsize=12)
        
        # Add statistics for Celeano
        cel_mean = np.mean(cel_flux)
        cel_std = np.std(cel_flux)
        cel_var = (cel_std / cel_mean) * 100
        ax1.text(0.02, 0.98, f'Mean Flux: {cel_mean:.0f}\nStd Dev: {cel_std:.0f}\nVariability: {cel_var:.3f}%\nData Points: {len(cel_flux)}', 
                transform=ax1.transAxes, verticalalignment='top',
                bbox=dict(boxstyle='round', facecolor='lightcyan', alpha=0.8))
    else:
        ax1.text(0.5, 0.5, "No data available for Celeano", 
                 ha='center', va='center', fontsize=14)
    
    # Plot Asterope
    if len(ast_time) > 0:
        # Normalize time to start from 0
        ast_time_norm = ast_time - ast_time.min()
        
        ax2.plot(ast_time_norm, ast_flux, 'g-', linewidth=1.5, alpha=0.8, label='Asterope')
        ax2.scatter(ast_time_norm, ast_flux, c='purple', s=5, alpha=0.7)
        ax2.set_title('Asterope Light Curve', fontsize=14, fontweight='bold')
        ax2.set_xlabel('Time (Days from start)', fontsize=12)
        ax2.set_ylabel('Flux (ADU)', fontsize=12)
        ax2.grid(True, alpha=0.3)
        ax2.legend()
        
        # Add statistics for Asterope
        ast_mean = np.mean(ast_flux)
        ast_std = np.std(ast_flux)
        ast_var = (ast_std / ast_mean) * 100
        ax2.text(0.02, 0.98, f'Mean Flux: {ast_mean:.0f}\nStd Dev: {ast_std:.0f}\nVariability: {ast_var:.3f}%\nData Points: {len(ast_flux)}', 
                transform=ax2.transAxes, verticalalignment='top',
                bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.8))
    else:
        ax2.text(0.5, 0.5, "No data available for Asterope", 
                 ha='center', va='center', fontsize=14)
    
    plt.tight_layout()
    plt.show()
    
    # Print summary statistics
    print("=== LIGHT CURVE ANALYSIS ===")
    if len(cel_time) > 0:
        print("\nCeleano:")
        print(f"  Data points: {len(cel_flux)}")
        print(f"  Time span: {cel_time_norm.max():.3f} days")
        print(f"  Time range: {cel_time.min():.3f} - {cel_time.max():.3f} (HJD)")
        print(f"  Mean flux: {cel_mean:.0f} ADU")
        print(f"  Flux range: {cel_flux.min():.0f} - {cel_flux.max():.0f} ADU")
        print(f"  Variability: {cel_var:.3f}%")
    
    if len(ast_time) > 0:
        print("\nAsterope:")
        print(f"  Data points: {len(ast_flux)}")
        print(f"  Time span: {ast_time_norm.max():.3f} days")
        print(f"  Time range: {ast_time.min():.3f} - {ast_time.max():.3f} (HJD)")
        print(f"  Mean flux: {ast_mean:.0f} ADU")
        print(f"  Flux range: {ast_flux.min():.0f} - {ast_flux.max():.0f} ADU")
        print(f"  Variability: {ast_var:.3f}%")

# Example usage with your file paths
celeano_file = './data/celeano-lightkurve-ts.csv'
asterope_file = './data/Asterope-lightkurve-ts.csv'

# Call the function to read and plot the data
plot_star_light_curves(celeano_file, asterope_file)
```
