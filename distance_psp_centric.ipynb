# PSP centric

This notebook creates a file where all the other planets and the sun are all calculated relative to the Parker Solar Probe

```python
import pandas as pd
import numpy as np

# Read the CSV file
data = pd.read_csv('./data/planet_and_psp_positions.csv')

# Function to calculate distance between two points (x1, y1, z1) and (x2, y2, z2)
def calculate_distance(x1, y1, z1, x2, y2, z2):
    return np.sqrt((x2 - x1)**2 + (y2 - y1)**2 + (z2 - z1)**2)

# Create an empty list to store new rows
distances_list = []

# Loop over each row in the data
for index, row in data.iterrows():
    psp_x = row['psp_x']
    psp_y = row['psp_y']
    psp_z = row['psp_z']
    
    # Calculate the distance to each planet and the Sun (Sun is at (0,0,0))
    distances_list.append({
        'date': row['date'],
        'distance_to_mercury': calculate_distance(psp_x, psp_y, psp_z, row['mercury_x'], row['mercury_y'], row['mercury_z']),
        'distance_to_venus': calculate_distance(psp_x, psp_y, psp_z, row['venus_x'], row['venus_y'], row['venus_z']),
        'distance_to_earth': calculate_distance(psp_x, psp_y, psp_z, row['earth_x'], row['earth_y'], row['earth_z']),
        'distance_to_sun': calculate_distance(psp_x, psp_y, psp_z, 0, 0, 0)  # Sun is at the origin (0,0,0)
    })

# Convert the list of dictionaries to a DataFrame
distances = pd.DataFrame(distances_list)

# Save the distances DataFrame to a new CSV file
distances.to_csv('./data/psp_distances.csv', index=False)

# Display the first few rows of the distances DataFrame
print(distances.head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the distances DataFrame (from the CSV generated previously)
distances = pd.read_csv('./data/psp_distances.csv')

# Plotting the distance to each planet
plt.figure(figsize=(12, 6))

# Plot each planet's distance with different styles and thickness
plt.plot(distances['date'], distances['distance_to_mercury'], label='Mercury', color='orange', linestyle='-', linewidth=1)
plt.plot(distances['date'], distances['distance_to_earth'], label='Earth', color='lightblue', linestyle='--', linewidth=1)

# Sun and Venus with more prominent styling
plt.plot(distances['date'], distances['distance_to_sun'], label='Sun', color='yellow', linestyle='-', linewidth=3, alpha=1.0)
plt.plot(distances['date'], distances['distance_to_venus'], label='Venus', color='lightgreen', linestyle='-', linewidth=3, alpha=1.0)

# Customize the plot
plt.xlabel('Date', fontsize=12)
plt.ylabel('Distance (AU)', fontsize=12)
plt.title('Distance from Parker Solar Probe to Planets and Sun', fontsize=14)
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for readability
plt.legend()
plt.grid(True)

# Show the plot
plt.tight_layout()  # Ensure everything fits in the plot area
plt.show()
```
